```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('diamonds.csv')
df = df.drop('Unnamed: 0',axis=1)
df
```

```python
cut = df.groupby('cut').mean().reset_index()
cut.plot(x='cut',y='price',kind='bar')
plt.ticklabel_format(style='plain', axis='y')
```

```python
cut.plot(x='cut',y='carat',kind='bar')
plt.ticklabel_format(style='plain', axis='y')
```

```python
color = df.groupby('color').mean().reset_index()
color.plot(x='color',y='price',kind='bar')
```

```python
color.plot(x='color',y='carat',kind='bar')
```

```python
clarity = df.groupby('clarity').mean().reset_index()
clarity.plot(x='clarity',y='price',kind='bar')
```

```python
clarity.plot(x='clarity',y='carat',kind='bar')
```
